```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
```

```python
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis, marker = "*", color="blue", linewidth=2, label="Boston")
plt.ylabel('Fare ($)')
plt.xlabel("Date")
plt.title("Pyber Fare by Month")
plt.legend()
plt.grid()
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, label="Boston", marker ="D", linewidth=2,color="green")
ax.set_ylabel("Fare ($)")
ax.set_xlabel("Date")
ax.set_title("Pyber Fare by Month")
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis,"o")
```

```python
plt.scatter(y_axis,x_axis,label="Chicago", color="red")
plt.title("Pyber Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0,50)
plt.gca().invert_yaxis()
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis,[i*5 for i in y_axis],color="skyblue",linewidth=2,label="boston")
ax.legend()
ax.set_title("Pyber Plot")
ax.set_xlabel("Date")
ax.set_ylabel("Fare")
```
